```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
"""
    Horn & Schunck Optical Flow
    
        img1, img2: image
        aplha: lambda
        iteration: iteration times
"""
def HS_optical_flow(img1, img2, alpha=1.0, iteration=64):
    img1 = img1.astype(np.float64)
    img2 = img2.astype(np.float64)
    
    # img gradients
    kernelX = np.array([[-1, 1], [-1, 1]]) * .25
    kernelY = np.array([[-1, -1], [1, 1]]) * .25
    kernelT = np.array([[1, 1], [1, 1]]) * .25
    
    Ix = cv2.filter2D(img1, -1, kernelX) + cv2.filter2D(img2, -1, kernelX)
    Iy = cv2.filter2D(img1, -1, kernelY) + cv2.filter2D(img2, -1, kernelY)
    It = cv2.filter2D(img2, -1, kernelT) - cv2.filter2D(img1, -1, kernelT)
    
    # initialize flow field
    U = np.zeros(img1.shape)
    V = np.zeros(img1.shape)
    
    kernelAvg = np.array([
        [0,   1/4,   0],
        [1/4,   0, 1/4],
        [0,   1/4,   0]
    ], dtype=np.float64)
    
    """
    kernelAvg = np.array([
        [1/12, 1/6, 1/12],
        [1/6,    0,  1/6],
        [1/12, 1/6, 1/12]
    ], dtype=np.float64)
    
    kernelAvg = np.array([
        [1/8, 1/8, 1/8],
        [1/8,   0, 1/8],
        [1/8, 1/8, 1/8]
    ], dtype=np.float64)
    
    """
    
    for _ in range(iteration):
        uAvg = cv2.filter2D(U, -1, kernelAvg)
        vAvg = cv2.filter2D(V, -1, kernelAvg)
        temp = ((Ix*uAvg + Iy*vAvg + It) / (1/alpha + Ix*Ix + Iy*Iy) )
        U = uAvg - temp * Ix
        V = vAvg - temp * Iy
        
    return U, V
    
```

```python
lena = cv2.imread("./lena.bmp", 0) #1: color, 0: gray, -1:unchanged
synthetic_lena = np.zeros(lena.shape).astype(np.int32)
synthetic_lena[1:, 1:] = lena[:-1, :-1]
print(lena)
print(synthetic_lena)
plt.imshow(lena, cmap="gray", vmin=0, vmax=255)
plt.show()
plt.imshow(synthetic_lena, cmap="gray", vmin=0, vmax=255)
plt.show()
```

```python
X = np.arange(lena.shape[0])
Y = np.arange(lena.shape[1])
```

```python
[U, V] = HS_optical_flow(lena, synthetic_lena, alpha = 0.1)
#Where X, Y define the arrow locations, U, V define the arrow directions, and C optionally sets the color.
fig, ax = plt.subplots()
#plt.gca().invert_yaxis()
#ax.invert_yaxis()
q = ax.quiver(Y, X, V, U,angles='xy')
ax.quiverkey(q, X=0.4, Y=1.1, U=10,
             label='HS', labelpos='E')
ax.invert_yaxis()
plt.gca().set_aspect('equal', adjustable='box')
plt.savefig(fname='lena01.png', dpi=1000)
plt.show()
```

```python
# Opencv plot dense optical flow
# https://docs.opencv.org/4.1.2/d4/dee/tutorial_optical_flow.html

hsv = np.zeros((512, 512, 3), dtype=np.uint8)
hsv[..., 1] = 255

mag, ang = cv2.cartToPolar(U, V)
hsv[..., 0] = ang * 180 / np.pi / 2
hsv[..., 2] = cv2.normalize(mag, None, 0, 255, cv2.NORM_MINMAX)
bgr = cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR)
plt.imshow(cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB), vmin=0, vmax=255)
cv2.imwrite("lena_dense_01.png", bgr)
plt.show()
```
