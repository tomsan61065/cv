```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import sys
import 
```

```python
img = cv2.imread("./lena.bmp", 0)
plt.imshow(img, cmap="gray")
plt.show()
```

```python
octagon = np.ones((5, 5))
octagon[0][0] = 0;
octagon[4][4] = 0;
octagon[0][4] = 0;
octagon[4][0] = 0;
print(octagon)
```

```python
def dilation(img, octagon):
    dilationImg = np.zeros((512, 512))
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            tempMax = 0
            for m in range(5):
                for n in range(5):
                    if i + m - 2 >= 0 and i + m - 2 < 512\
                    and j + n - 2 >= 0 and j + n - 2 < 512:
                        tempMax = max(tempMax, img[i + m - 2][j + n - 2]); # 5 * 5
            dilationImg[i][j] = tempMax
    return dilationImg
dilationImg = dilation(img, octagon)
print(dilationImg.shape)
plt.imshow(dilationImg, cmap="gray")
plt.show()
```

```python
def erosion(img, octagon):
    erosionImg = np.zeros((512, 512))
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            tempMin = 255
            for m in range(5):
                for n in range(5):
                    if i + m - 2 >= 0 and i + m - 2 < 512\
                    and j + n - 2 >= 0 and j + n - 2 < 512:
                        tempMin = min(tempMin, img[i + m - 2][j + n - 2]); # 5 * 5
            erosionImg[i][j] = tempMin
    return erosionImg
erosionImg = erosion(img, octagon)
plt.imshow(erosionImg, cmap="gray")
plt.show()
```

```python
def opening(img, octagon):
    ero = erosion(img, octagon)
    return dilation(ero, octagon)

openingImg = opening(img, octagon)
plt.imshow(openingImg, cmap="gray")
plt.show()
```

```python
def closing(img, octagon):
    dil = dilation(img, octagon)
    return erosion(dil, octagon)

closingImg = closing(img, octagon)
plt.imshow(closingImg, cmap="gray")
plt.show()
```

```python
#https://www.cnblogs.com/qrlozte/p/5340216.html
def SNR(originImg, noiseImg):
    originImg = originImg.astype(np.float64)
    originImg /= 255.0
    noiseImg = noiseImg.astype(np.float64)
    noiseImg /= 255.0
    return 10*np.log10( np.sum(np.power(originImg, 2)) / np.sum(np.power(originImg - noiseImg, 2)) )
```

```python
print(SNR(img, closingImg))
```

```python
img2 = cv2.imread("./median_5x5.bmp", 0)
print(SNR(img, img2))
```

```python
def numpy_SNR(origianl_waveform, target_waveform):
    # 单位 dB
    origianl_waveform = origianl_waveform.astype(np.float64)
    target_waveform = target_waveform.astype(np.float64)
    origianl_waveform /= 255.0
    target_waveform /= 255.0
    signal = np.sum(origianl_waveform ** 2)
    noise = np.sum((origianl_waveform - target_waveform) ** 2)
    snr = 10 * np.log10(signal / noise)
    return snr
```

```python
print(numpy_SNR(img, img2))
```

```python
def bookSNR(originImg, noiseImg):
    originImg = originImg.astype(np.float64)
    originImg /= 255.0
    noiseImg = noiseImg.astype(np.float64)
    noiseImg /= 255.0
    
    return 10 * np.log10( np.var(originImg) / np.var(originImg - noiseImg) )
    #return 20 * np.log( np.sqrt(np.var(originImg)) / np.sqrt(np.var(noiseImg)) )
```

```python
print(bookSNR(img, img2))
```

```python
def generateGaussianNoise(img, amplitude=1):
    noise = np.random.normal(size=img.shape[0]*img.shape[1])
    noise = np.reshape(noise, (img.shape[0], img.shape[1]))
    return (noise * amplitude) + img
```

```python
G10Img = generateGaussianNoise(img, 10)
plt.imshow(G10Img, cmap="gray")
plt.show()
print(bookSNR(img, G10Img))
G30Img = generateGaussianNoise(img, 30)
plt.imshow(G30Img, cmap="gray")
plt.show()
print(bookSNR(img, G30Img))
```

```python
def generateSaltAndPepperNoise(img, threshold):
    SAPImg = np.copy(img)
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            val = np.random.uniform(size=1)
            if val <= threshold:
                SAPImg[i][j] = 0
            elif val >= 1 - threshold:
                SAPImg[i][j] = 255
    return SAPImg
```

```python
SAP10Img = generateSaltAndPepperNoise(img, 0.1)
plt.imshow(SAP10Img, cmap="gray")
plt.show()
print(bookSNR(img, SAP10Img))
SAP5Img = generateSaltAndPepperNoise(img, 0.05)
plt.imshow(SAP5Img, cmap="gray")
plt.show()
print(bookSNR(img, SAP5Img))
```

```python
def Box(img, val):
    # 3 -> 1, 5 -> 2
    border = val//2
    borderImg = cv2.copyMakeBorder(img, border, border, border, border,cv2.BORDER_REPLICATE)
    resultImg = np.copy(img)
    for i in range(img.shape[0]):
        for j in range(img.shape[0]):
            resultImg[i][j] = np.mean(borderImg[i:i+val, j:j+val])
    return resultImg
```

```python
def median(img, val):
    # 3 -> 1, 5 -> 2
    border = val//2
    borderImg = cv2.copyMakeBorder(img, border, border, border, border,cv2.BORDER_REPLICATE)
    resultImg = np.copy(img)
    for i in range(img.shape[0]):
        for j in range(img.shape[0]):
            resultImg[i][j] = np.median(borderImg[i:i+val, j:j+val])
    return resultImg
```
