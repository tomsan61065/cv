```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import sys
```

```python
img = cv2.imread("./lena.bmp", 0)
plt.imshow(img, cmap="gray")
plt.show()
```

```python
#https://en.wikipedia.org/wiki/Discrete_Laplace_operator
def Laplacian(img, isEight=False):
    img = img.astype(np.float64)
    border = 1
    borderImg = cv2.copyMakeBorder(img, border, border, border, border,cv2.BORDER_REPLICATE)
    
    kernal = np.array([
        [0, 1, 0],
        [1,-4, 1], 
        [0, 1, 0]
    ])
    
    kernal2 = np.array([
        [1, 1, 1],
        [1,-8, 1], 
        [1, 1, 1]
    ])
    
    ResultImg = np.zeros(img.shape)
    #print(GxImg.shape)
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            val = 0.0
            for m in range(3):
                for n in range(3):
                    if(isEight):
                        val += borderImg[i + m][j + n] * kernal2[m][n] // 3
                    else:
                        val += borderImg[i + m][j + n] * kernal[m][n]
            ResultImg[i][j] = val
    
    
    
    return ResultImg
```

```python
LaplacianImg = Laplacian(img)
LaplacianImg[LaplacianImg >= 15] = 255
LaplacianImg[LaplacianImg < 15] = 0
LaplacianImg /= 255.0
LaplacianImg = 1 - LaplacianImg
plt.imshow(LaplacianImg, cmap="gray")
plt.show()

LaplacianImg = Laplacian(img, True)
LaplacianImg[LaplacianImg >= 15] = 255
LaplacianImg[LaplacianImg < 15] = 0
LaplacianImg /= 255.0
LaplacianImg = 1 - LaplacianImg
plt.imshow(LaplacianImg, cmap="gray")
plt.show()
```

```python
#https://stackoverflow.com/questions/48319918/whats-the-theory-behind-computing-variance-of-an-image
def minimumVarianceLaplacian(img):
    img = img.astype(np.float64)
    border = 1
    borderImg = cv2.copyMakeBorder(img, border, border, border, border,cv2.BORDER_REPLICATE)
    
    kernal = np.array([
        [ 2, -1,  2],
        [-1, -4, -1], 
        [ 2, -1,  2]
    ])
    
    
    ResultImg = np.zeros(img.shape)
    #print(GxImg.shape)
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            val = 0.0
            for m in range(3):
                for n in range(3):
                    val += borderImg[i + m][j + n] * kernal[m][n] // 3
            ResultImg[i][j] = val
    
    return ResultImg
```

```python
MVLaplacianImg = minimumVarianceLaplacian(img)
MVLaplacianImg[MVLaplacianImg >= 20] = 255
MVLaplacianImg[MVLaplacianImg < 20] = 0
MVLaplacianImg /= 255.0
MVLaplacianImg = 1 - MVLaplacianImg
plt.imshow(MVLaplacianImg, cmap="gray")
plt.show()
```

```python
#https://en.wikipedia.org/wiki/Blob_detection#The_Laplacian_of_Gaussian
def LoG(img):
    img = img.astype(np.float64)
    border = 5
    borderImg = cv2.copyMakeBorder(img, border, border, border, border,cv2.BORDER_REPLICATE)
    
    kernal = np.array([
        [ 0,   0,   0,  -1,  -1,  -2,  -1,  -1,   0,   0,   0],
        [ 0,   0,  -2,  -4,  -8,  -9,  -8,  -4,  -2,   0,   0],
        [ 0,  -2,  -7, -15, -22, -23, -22, -15,  -7,  -2,   0],
        [-1,  -4, -15, -24, -14,  -1, -14, -24, -15,  -4,  -1],
        [-1,  -8, -22, -14,  52, 103,  52, -14, -22,  -8,  -1],
        [-2,  -9, -23,  -1, 103, 178, 103,  -1, -23,  -9,  -2],
        [-1,  -8, -22, -14,  52, 103,  52, -14, -22,  -8,  -1],
        [-1,  -4, -15, -24, -14,  -1, -14, -24, -15,  -4,  -1],
        [ 0,  -2,  -7, -15, -22, -23, -22, -15,  -7,  -2,   0],
        [ 0,   0,  -2,  -4,  -8,  -9,  -8,  -4,  -2,   0,   0],
        [ 0,   0,   0,  -1,  -1,  -2,  -1,  -1,   0,   0,   0]
    ])
    
    
    ResultImg = np.zeros(img.shape)
    #print(GxImg.shape)
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            val = 0.0
            for m in range(11):
                for n in range(11):
                    val += borderImg[i + m][j + n] * kernal[m][n]
            ResultImg[i][j] = val
    
    return ResultImg
```

```python
LoGImg = LoG(img)
LoGImg[LoGImg < 3000] = 0
LoGImg[LoGImg >= 3000] = 255
LoGImg /= 255
LoGImg = 1 - LoGImg
plt.imshow(LoGImg, cmap="gray")
plt.show()
```

```python
#https://en.wikipedia.org/wiki/Robinson_compass_mask
def Robinson(img):
    img = img.astype(np.float64)
    border = 1
    borderImg = cv2.copyMakeBorder(img, border, border, border, border,cv2.BORDER_REPLICATE)
    G = []
    G.append(np.array([
        [-1,  0,  1],
        [-2,  0,  2], 
        [-1,  0,  1]
    ]))
    G.append(np.array([
        [ 0,  1,  2],
        [-1,  0,  1], 
        [-2, -1,  0]
    ]))
    G.append(np.array([
        [ 1,  2,  1],
        [ 0,  0,  0], 
        [-1, -2, -1]
    ]))
    G.append(np.array([
        [ 2,  1,  0],
        [ 1,  0, -1], 
        [ 0, -1, -2]
    ]))
    
    G.append(np.array([
        [ 1,  0, -1],
        [ 2,  0, -2], 
        [ 1,  0, -1]
    ]))
    G.append(np.array([
        [ 0, -1, -2],
        [ 1,  0, -1], 
        [ 2,  1,  0]
    ]))
    G.append(np.array([
        [-1, -2, -1],
        [ 0,  0,  0], 
        [ 1,  2,  1]
    ]))
    G.append(np.array([
        [-2, -1,  0],
        [-1,  0,  1], 
        [ 0,  1,  2]
    ]))
    G = np.array(G)
    resultImg = np.zeros(img.shape)
    #print(GxImg.shape)
    for z in range(8):
        for i in range(img.shape[0]):
            for j in range(img.shape[1]):
                val = 0.0
                for m in range(3):
                    for n in range(3):
                        val += borderImg[i + m][j + n] * G[z][m][n]
                resultImg[i][j] = max(val, resultImg[i][j])
    
    return resultImg
```

```python
RobinsonImg = Robinson(img)
RobinsonImg[RobinsonImg >= 43] = 255
RobinsonImg[RobinsonImg < 43] = 0
RobinsonImg /= 255.0
RobinsonImg = 1 - RobinsonImg
plt.imshow(RobinsonImg, cmap="gray")
plt.show()
```

```python
#https://en.wikipedia.org/wiki/Robinson_compass_mask
def Nevatia_Babu(img):
    img = img.astype(np.float64)
    border = 2
    borderImg = cv2.copyMakeBorder(img, border, border, border, border,cv2.BORDER_REPLICATE)
    G = []
    G.append(np.array([
        [ 100, 100, 100, 100, 100],
        [ 100, 100, 100, 100, 100],
        [   0,   0,   0,   0,   0],
        [-100,-100,-100,-100,-100],
        [-100,-100,-100,-100,-100]
    ]))
    G.append(np.array([
        [ 100, 100, 100, 100, 100],
        [ 100, 100, 100,  78, -32],
        [ 100,  92,   0, -92,-100],
        [  32, -78,-100,-100,-100],
        [-100,-100,-100,-100,-100]
    ]))
    G.append(np.array([
        [ 100, 100, 100,  32,-100],
        [ 100, 100, 100, -78,-100],
        [ 100, 100,   0,-100,-100],
        [ 100,  78, -92,-100,-100],
        [ 100, -32,-100,-100,-100]
    ]))
    
    G.append(np.array([
        [-100,-100,   0, 100, 100],
        [-100,-100,   0, 100, 100],
        [-100,-100,   0, 100, 100],
        [-100,-100,   0, 100, 100],
        [-100,-100,   0, 100, 100]
    ]))
    G.append(np.array([
        [-100,  32, 100, 100, 100],
        [-100, -78,  92, 100, 100],
        [-100,-100,   0, 100, 100],
        [-100,-100, -92,  78, 100],
        [-100,-100,-100, -32, 100]
    ]))
    G.append(np.array([
        [ 100, 100, 100, 100, 100],
        [ -32,  78, 100, 100, 100],
        [-100, -92,   0,  92, 100],
        [-100,-100,-100, -78,  32],
        [-100,-100,-100,-100,-100]
    ]))
    G = np.array(G)
    resultImg = np.zeros(img.shape)
    #print(GxImg.shape)
    for z in range(6):
        for i in range(img.shape[0]):
            for j in range(img.shape[1]):
                val = 0.0
                for m in range(5):
                    for n in range(5):
                        val += borderImg[i + m][j + n] * G[z][m][n]
                resultImg[i][j] = max(val, resultImg[i][j])
    
    return resultImg
```
