```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("./lena.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
M = cv2.getRotationMatrix2D((img.shape[1]//2, img.shape[0]//2), -45, 1.0)
rotateImg = cv2.warpAffine(img, M, (img.shape[1], img.shape[0]), flags=cv2.INTER_NEAREST)
plt.imshow(rotateImg)
plt.show()
rotateImg = cv2.cvtColor(rotateImg, cv2.COLOR_RGB2BGR)
cv2.imwrite("rotateImg.png", rotateImg)
```

```python
resizeImg = cv2.resize(img, (img.shape[0]//2, img.shape[1]//2), interpolation=cv2.INTER_CUBIC)
plt.imshow(resizeImg)
plt.show()
resizeImg = cv2.cvtColor(resizeImg, cv2.COLOR_RGB2BGR)
cv2.imwrite("resizeImg.png", resizeImg)
```

```python
grayImg = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(grayImg, cmap="gray")
plt.show()
```

```python
binaryImg = np.zeros((512, 512))
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        if grayImg[i][j] > 128:
            binaryImg[i][j] = 255
        else:
            binaryImg[i][j] = 0
plt.imshow(binaryImg, cmap="gray")
plt.show()
#cv2.imwrite("binaryImg.png", binaryImg)
```
